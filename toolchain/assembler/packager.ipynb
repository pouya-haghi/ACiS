```python
NUM_COL = 2
DWIDTH = 512
IWIDTH = 32
NUM_INS_PER_BEAT = int(DWIDTH/IWIDTH)
def packager(binary_filename, packaged_filename, num_col):
    with open(packaged_filename, 'w') as g:
        with open(binary_filename) as f:
            lines = f.readlines()
            for i in range(len(lines)):
                this_line = lines[i]
                integers = [int(this_line[step:step+8], 2)  # Extract 8 bits at a time and convert to integer
                for step in range(0, 32, 8)]  # Iterate in steps of 8 bits
                for col in range(num_col):
                    g.write(str(integers[3])+'\n')
                    g.write(str(integers[2])+'\n')
                    g.write(str(integers[1])+'\n')
                    g.write(str(integers[0])+'\n')
                for remaining_col in range(NUM_INS_PER_BEAT - num_col):
                    g.write(str(0)+'\n')
                    g.write(str(0)+'\n')
                    g.write(str(0)+'\n')
                    g.write(str(0)+'\n')
```

```python
packager('./Assembly_code/sample_test_intACiS.bin', './Assembly_code/sample_test_intACiS.pkg', NUM_COL)
```
